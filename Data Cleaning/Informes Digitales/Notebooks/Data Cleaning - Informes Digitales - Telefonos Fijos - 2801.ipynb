### 2.1.1) Análisis del campo "Telefonos propios"

```python
df_string_split_1 = df_reduced['Telefonos propios'].str.split(regex_tels, expand=True)
```

```python
df_string_split_1.shape
```

Bien, se mantiene el mismo número de filas.

```python
df_string_split_1.head(15)
```

Generamos una máscara para identificar en qué celdas del dataframe tengo esa regex

```python
df_masked_1 = df_string_split_1.mask(~df_string_split_1.apply (lambda x: x.str.contains(regex_tels,na=False)))
```

```python
df_masked_1.head(15)
```

```python
df_masked_1.shape
```

Dropeamos todas las filas completamente duplicadas que puedan existir

```python
#df_masked_1 = df_masked_1.drop_duplicates()
```

```python
df_masked_1.shape
```

```python
df_masked_1 ['Telefonos_propios_list'] = pd.Series(df_masked_1.values.tolist(), index=df_masked_1.index)
```

# 3) Diccionarios

Vamos a generar un diccionario por campo del dataset que contenga telefonos fijos.

## 3.1) Campo "Tel. de contacto e email"

Para generar el diccionario vamos a unir primero con el dataframe original para quedarnos con los DNIs. Posteriormente, vamos a unir con el resto de los campos de interés.

```python
df_tel_propios_joined_1 = df_telefonos_propios.join (df_reduced['DNI'])
```

```python
df_tel_propios_joined_1
```

Reordenamos las columnas

```python
df_tel_propios_joined_1 = df_tel_propios_joined_1[['DNI', 'Telefonos_propios_list']]
```

```python
df_tel_propios_joined_1.head()
```

### 3.1.1) Agrupamos por DNI

Vamos a agrupar por DNI

```python
df_grouped_1 = df_tel_propios_joined_1.groupby('DNI').agg({'Telefonos_propios_list': 'sum'})
```

```python
df_grouped_1
```

Reordenamos las columnas

### 3.1.2) Eliminación de NaNs

Vamos a eliminar todos los NaNs de las listas

```python
df_grouped_1['Telefonos_propios_list'] = [[x for x in df_grouped_1['Telefonos_propios_list'] [i] if pd.notnull(x)] for i in range (len(df_grouped_1['Telefonos_propios_list']))]
```

```python
df_grouped_1
```

### 3.1.3) Eliminación de Duplicados dentro de cada lista

Vamos a eliminar todos los duplicados qeu puedan aparecer dentro de cada lista de mails

```python
df_grouped_1['Telefonos_propios_list'] = [list(dict.fromkeys(df_grouped_1['Telefonos_propios_list'][i])) for i in range (len(df_grouped_1['Telefonos_propios_list']))]
```

```python
df_grouped_1
```

### 3.1.4) Conversión a Diccionarios

Vamos a convertir este dataframe a un diccionario para luego mergearlo

```python
dict_1 = df_grouped_1.to_dict()
```

```python
dict_1
```

```python
dict_1 = dict_1['Telefonos_propios_list']
```

Limpiamos el dataframe __df_infdig__, ya que dentro de DATO_A_ENVIAR tenemos 9 tipos de casos diferentes de datos:

- DNI 12345678
- DNI12345678
- DNIE1234678
- LC  1234567

- 00123456780 --> Es un CUIT.
- 00 12345678 0 --> Es un CUIT.
- DNI 00123456780 --> Dice DNI pero es un CUIT de una persona __jurídica__ (TO DO)

Para eso definimos 3 funciones:

- __Standard DNI__: Estandariza los DNI para 7 de los casos. Esta función debe ejecutarse en primer lugar.
- __DNI Index__: Estandariza para los casos remanentes de Standard DNI.
- __DNI Zero__: Elimina los ceros iniciales de ciertos CUITs para que coincidan con los DNI.

```python
def standard_dni (string_dni):
    """DNI separation by each case.
    """
    
    if 'DNI ' in string_dni:
        dni = string_dni.split('DNI ')[1]
        
    elif 'DNIE' in string_dni:
        dni = string_dni.split('DNIE')[1]
    
    elif 'DNI' in string_dni:
        dni = string_dni.split('DNI')[1]
    
    elif 'LC  ' in string_dni:
        dni = string_dni.split('LC  ')[1]
    
    elif 'LE  ' in string_dni:
        dni = string_dni.split('LE  ')[1]
    
    elif 'CUIT' in string_dni:
        dni = string_dni.split('CUIT')[1]
        
    else:
        dni = string_dni
        
    return dni
```

```python
def dni_characters (string_dni):
    
    if ' ' in string_dni:
        dni = string_dni.replace(' ','')
        
    elif '-' in string_dni:
        dni = string_dni.replace('-','')

    else:
        dni = string_dni
    
    return dni
```

```python
def cuit (string_dni):
    
    if ((string_dni.startswith('20')) & (len(string_dni)>8)):
        cuit = string_dni [2:-1]
        
    elif ((string_dni.startswith('23')) & (len(string_dni)>8)):
        cuit = string_dni [2:-1]
        
    elif ((string_dni.startswith('24')) & (len(string_dni)>8)):
        cuit = string_dni [2:-1]

    elif ((string_dni.startswith('27')) & (len(string_dni)>8)):
        cuit = string_dni [2:-1]

    else:
        cuit = string_dni
        
    return cuit
```

```python
def dni_zero (string_dni):
    
    if string_dni.startswith('0'):
        dni = string_dni [1:]

    else:
        dni = string_dni
    
    return dni
```

```python
df_infdig['DNI'] = df_infdig['DNI'].astype(str)
```

En primer lugar, trimmeamos los espacios que sobran:

```python
df_infdig['DNI'] = [(df_infdig['DNI'][i].strip()) for i in range (df_infdig['DNI'].count())]
```

Se implementan estas tres funciones:

```python
df_infdig['DNI'] = [standard_dni(df_infdig['DNI'][i]) for i in range (df_infdig['DNI'].count())]
```

```python
df_infdig['DNI'] = [dni_characters(df_infdig['DNI'][i]) for i in range (df_infdig['DNI'].count())]
```

```python
df_infdig['DNI'] = [cuit(df_infdig['DNI'][i]) for i in range (df_infdig['DNI'].count())]
```

```python
df_infdig['DNI'] = [dni_zero(df_infdig['DNI'][i]) for i in range (df_infdig['DNI'].count())]
```
