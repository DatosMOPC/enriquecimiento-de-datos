## Dataframe reducido

```python
df_reduced = df [['Cuit','Nombre','TipoDocu','NumeDocu','Refno1','Area1','Prefij1','Refte1','Refno2','Area2','Prefij2','Refte2']]
```

```python
df_reduced ['TIPO_DOC'] = 'DNI'
```

```python
df_reduced ['NumeDocu'] = df_reduced['NumeDocu'].astype(str)
```

### GeneraciÃ³n de IDMOROSO

```python
df_reduced ['IDMOROSO'] = df_reduced ['TIPO_DOC'] + " " + df_reduced ['NumeDocu']
```

```python
df_reduced.head()
```

### ConcatenaciÃ³n de los campos de telÃ©fono

```python
df_reduced ['Area1'] = df_reduced['Area1'].astype(str)
df_reduced ['Area2'] = df_reduced['Area2'].astype(str)

df_reduced ['Prefij1'] = df_reduced['Prefij1'].astype(str)
df_reduced ['Prefij2'] = df_reduced['Prefij2'].astype(str)

df_reduced ['Refte1'] = df_reduced['Refte1'].astype(str)
df_reduced ['Refte2'] = df_reduced['Refte2'].astype(str)

df_reduced ['Cuit'] = df_reduced['Cuit'].astype(str)
```

```python
df_reduced.loc[df_reduced ['Area1'] == 'nan', 'Area1'] = np.nan
df_reduced.loc[df_reduced ['Area2'] == 'nan', 'Area2'] = np.nan

df_reduced.loc[df_reduced ['Prefij1'] == 'nan', 'Prefij1'] = np.nan
df_reduced.loc[df_reduced ['Prefij2'] == 'nan', 'Prefij2'] = np.nan

df_reduced.loc[df_reduced ['Refte1'] == 'nan', 'Refte1'] = np.nan
df_reduced.loc[df_reduced ['Refte2'] == 'nan', 'Refte2'] = np.nan

df_reduced.loc[df_reduced ['Cuit'] == 'nan', 'Cuit'] = np.nan
```

```python
df_reduced.head()
```

```python
df_reduced ['Area1'] = df_reduced ['Area1'].str.split('.').str[0]
df_reduced ['Prefij1'] = df_reduced ['Prefij1'].str.split('.').str[0]
df_reduced ['Refte1'] = df_reduced ['Refte1'].str.split('.').str[0]

df_reduced ['Area2'] = df_reduced ['Area2'].str.split('.').str[0]
df_reduced ['Prefij2'] = df_reduced ['Prefij2'].str.split('.').str[0]
df_reduced ['Refte2'] = df_reduced ['Refte2'].str.split('.').str[0]

df_reduced ['Cuit'] = df_reduced ['Cuit'].str.split('.').str[0]
```

```python
df_reduced.head()
```

```python
df_reduced ['TEL1'] = df_reduced ['Prefij1'] + df_reduced ['Area1'] + df_reduced ['Refte1']
df_reduced ['TEL2'] = df_reduced ['Prefij2'] + df_reduced ['Area2'] + df_reduced ['Refte2']
```

```python
df_reduced.head()
```

### Dropeamos y renombramos

```python
df_reduced = df_reduced.rename(columns = {'NumeDocu':'NRO_DOC','Cuit':'CUIT_CUIL'})
```

```python
df_reduced = df_reduced [['IDMOROSO','Nombre','TIPO_DOC','NRO_DOC','CUIT_CUIL','TEL1','TEL2']]
```

```python
df_reduced.head()
```

```python
df_melt = pd.melt(df_reduced, id_vars=['IDMOROSO','Nombre','TIPO_DOC','NRO_DOC','CUIT_CUIL'], value_vars=['TEL1','TEL2'],value_name='DATO')
```

```python
df_melt.head()
```

```python
df_melt.shape
```

```python
df_melt = df_melt.dropna(subset=['DATO'])
```

```python
df_melt.shape
```

### Propietario del Dato

En este caso, serÃ¡ "Ajeno_referente"

```python
df_clean = df_melt.assign(PROPIETARIO_DATO='Ajeno_referente')
```

```python
df_clean = df_clean.assign(PROVEEDOR_DATO='C_FRV')
```

```python
df_clean = df_clean.assign(TIPO='TEL.CELULAR')
```

```python
df_clean = df_clean.assign(SUBTIPO='PERSONAL')
```

```python
df_clean = df_clean.drop(columns=['variable'])
```

```python
df_clean.head()
```

# 9) Columna Score

El campo **Score** se construye en funciÃ³n de los datos obtenidos:

- Nivel 4: Dato que otorga la propia persona.
- Nivel 3: Dato que otorgan los clientes (los bancos).
- Nivel 2: Dato que otorga un proveedor externo (Informes Digitales en este caso), con subtipo "Personal".
- Nivel 1: Dato que otorga un proveedor externo, pero tiene como subtipo "Laboral".

```python
# Nivel 1: Laboral, proveedor externo.

df_clean.loc[(df_clean['SUBTIPO'] == 'LABORAL') , 'SCORE_DATO'] = 1
```

```python
# Nivel 2: Personal, proveedor externo.

df_clean.loc[(df_clean['SUBTIPO'] == 'PERSONAL') , 'SCORE_DATO'] = 2
```

```python
df_clean.head()
```

# 10) Columna Fecha

Para el campo FECHA_INGRESO_DATO, usamos la fecha en la que se hace la inserciÃ³n de los datos:

```python
fecha_ingreso = pd.to_datetime('today')
```

```python
df_clean = df_clean.assign(FECHA_INGRESO_DATO=fecha_ingreso)
```

```python
df_clean.head()
```
