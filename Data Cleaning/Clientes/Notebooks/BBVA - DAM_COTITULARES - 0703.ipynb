Como las listas tienen longitudes fijas, puedo seleccionar los elementos que quiero según índice:

- 16 si es un único cotitular

- Nombre Cotitular: [2], [17].
- Tipo Producto: [0], [15].
- DNI Cotitular: [1],[16]

### Nombre Cotitular

Generamos las condiciones:

- Si la lista tiene longitud 16, tomamos el nombre solamente del elemento 2.
- Si la lista tiene otra longitud, formamos una lista con los nombres en los elementos 2 y 17.

```python
df_reduced ['Nombre_cotitular'] = [(df_reduced ['Cotitulares'].iloc[i][2]) if len(df_reduced ['Cotitulares'].iloc[i]) == 16 else ([df_reduced ['Cotitulares'].iloc[i][2]
                                                                    ,df_reduced ['Cotitulares'].iloc[i][17]]) 
                                                                     for i in range ((df_reduced.shape[0]))]
```

```python
df_reduced = df_reduced.explode('Nombre_cotitular')
```

```python
df_reduced ['Nombre_cotitular'] = df_reduced ['Nombre_cotitular'].str.strip()
```

```python
df_reduced.head()
```

### Tipo de Producto

Generamos las condiciones:

- Si la lista tiene longitud 16, tomamos el nombre solamente del elemento 0.
- Si la lista tiene otra longitud, formamos una lista con los nombres en los elementos 0 y 15.

```python
df_reduced ['TIPO_PRODUCTO'] = [(df_reduced ['Cotitulares'].iloc[i][0]) if len(df_reduced ['Cotitulares'].iloc[i]) == 16 else ([df_reduced ['Cotitulares'].iloc[i][0]
                                                                    ,df_reduced ['Cotitulares'].iloc[i][15]]) 
                                                                     for i in range ((df_reduced.shape[0]))]
```

```python
df_reduced = df_reduced.explode('TIPO_PRODUCTO')
```

```python
df_reduced ['TIPO_PRODUCTO'] = df_reduced ['TIPO_PRODUCTO'].str.strip()
```

```python
df_reduced.head()
```

### DNI Cotitular

```python
df_reduced = df_reduced.explode('DNI_COTITULAR')
```
